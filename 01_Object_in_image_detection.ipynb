```python
import cv2
import cvzone
from ultralytics import YOLO

import matplotlib.pyplot as plt
```

## Creating our model

```python
model = YOLO('yolov8n.pt')  # Ensure correct filename
results = model('/kaggle/input/yolo-dataset/New folder/john-kovacich-ld-19Zng4Ac-unsplash.jpg', show=True)

# Get the image with bounding boxes
annotated_img = results[0].plot()  # This returns a numpy array (BGR format)

# Convert BGR to RGB for matplotlib
annotated_img = cv2.cvtColor(annotated_img, cv2.COLOR_BGR2RGB)

# Show the result inline
plt.figure(figsize=(10, 8))
plt.imshow(annotated_img)
plt.axis('off')
plt.show()
```
